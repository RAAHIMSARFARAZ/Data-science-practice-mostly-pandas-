```python
import pandas as pd
df = pd.read_csv('/content/game_details.csv')
# shows the first four rows of the Dataset
df.head()
```

```python
# shows the last four rows of the Dataset.
df.tail()
```

```python
# shows you the Total entries, index ranges, columns with non-null counts and type
df.info()
```

```python
# Query lets you get a specific column name and get its specific price or in this case metacritic score
df.query('platform == "PS4" & metacritic_score >= 19')
```

```python
# shows you the sum of null values in the columns
df.isnull().sum()
```

```python
# shows you the column names that hold null values
df.columns[df.isnull().any()]
```

```python
# gives you any five samples from the Dataset
df.sample(5)
```

```python
# drops or deletes the null values in the dataset
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# shows any duplicated values or columns in the Data set
df[df.duplicated()]
```

```python
# find correlations between different values
df['metacritic_score'].corr(df['playstation_score'])
```

```python
# displays the mean, count, standards, minimum etc in the whole dataset
df.describe()
```

```python
# counts how many games are at origin or game name
df['game_name'].value_counts()
```

```python
# returns the mean of all the games's metacritic score
df.groupby('game_name')['metacritic_score'].mean()
```

```python
# rounds up the values in a certain column
df['metacritic_score'].apply(round)
```

```python
# converts the datatype in the dataset to a number with a decimal
df['metacritic_score'] = df['metacritic_score'].astype(float)
```

```python
# Shows the highest values in the dataset
df.nlargest(5, 'metacritic_score')
```

```python
# Shows the smallest values in the Dataset
df.nsmallest(5, 'metacritic_score')
```

```python
# saves the cleaned data to a separate csv file
df.to_csv('game_details_clean.csv', index=False)
```
